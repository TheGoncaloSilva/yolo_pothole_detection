```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Plot YOLOv11n patience 10 vs patience 100

```python
# Initialize lists to store the data
epochs = []
mAP50_95 = []
mAP50 = []
labels = []

# Filter the trainingResults to include only the specified models
patience_results = [trainingResults.get("yolov11n_250_epochs_10_patience_reduced_dataset"), trainingResults.get("yolov11n_250_epochs_100_patience_reduced_dataset")]

print(patience_results[0])
print(patience_results[1])
# Extract the data for the specified models
for model_info in patience_results:
    if model_info and "path" in model_info:
        df = pd.read_csv(model_info["path"])
        epochs.append(df['epoch'])
        mAP50_95.append(df['metrics/mAP50-95(B)'])
        mAP50.append(df['metrics/mAP50(B)'])

        labels.append(model_info["model"] + " " + str(model_info["epochs"]) + " epochs " + str(model_info["patience"]) + " patience")

# Plot the mAP metrics
plt.figure(figsize=(20, 10))
for i in range(len(epochs)):
    plt.plot(epochs[i], mAP50_95[i], label=labels[i])
    plt.plot(epochs[i], mAP50[i], label=labels[i] + " mAP50")
plt.xlabel('Epochs')
plt.ylabel('mAP50-95')
plt.title('mAP50-95')
plt.legend()
```

```python
all_epochs = []
all_dfl = []
all_class = []
all_box= []

# Filter the trainingResults to include only the specified models
version_results = [trainingResults.get("yolov5nu_250_epochs_10_patience"), trainingResults.get("yolov11n_250_epochs_100_patience_reduced_dataset"), trainingResults.get("yolov11s_250_epochs_10_patience_reduced_dataset")]

# Determine the common range of epochs
min_epoch = float('inf')
max_epoch = float('-inf')
for model_info in version_results:
    if model_info and "path" in model_info:
        df = pd.read_csv(model_info["path"])
        min_epoch = min(min_epoch, df['epoch'].min())
        max_epoch = max(max_epoch, df['epoch'].max())

# Create a common set of epochs
common_epochs = np.arange(min_epoch, max_epoch + 1)

# Extract the data for the specified models and interpolate to the common set of epochs
for model_info in version_results:
    if model_info and "path" in model_info:
        df = pd.read_csv(model_info["path"])
        epochs = df['epoch']
        dfl = df['val/dfl_loss']
        clss = df['val/cls_loss']
        box = df['val/box_loss']
        
        # Interpolate the losses to the common set of epochs
        dfl_interp = np.interp(common_epochs, epochs, dfl)
        clss_interp = np.interp(common_epochs, epochs, clss)
        box_interp = np.interp(common_epochs, epochs, box)
        
        all_epochs.append(common_epochs)
        all_dfl.append(dfl_interp)
        all_class.append(clss_interp)
        all_box.append(box_interp)

        labels.append(model_info["model"] + " " + str(model_info["epochs"]) + " epochs " + str(model_info["patience"]) + " patience")

# Plot the loss metrics
plt.figure(figsize=(20, 10))
for i in range(len(all_epochs)):
    plt.plot(all_epochs[i], all_dfl[i], label=labels[i] + " validation DFL Loss")
    plt.plot(all_epochs[i], all_class[i], label=labels[i] + " validation Class Loss")
    plt.plot(all_epochs[i], all_box[i], label=labels[i] + " validation Box Loss")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Metrics over Epochs')
plt.legend()
plt.show()
```
